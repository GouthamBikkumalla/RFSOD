```python
import os
import cv2
import matplotlib.pyplot as plt
import pandas as pd

train_df=pd.read_csv('/content/drive/MyDrive/DL_Project/train/_annotations.csv')
test_df=pd.read_csv('/content/drive/MyDrive/DL_Project/test/_annotations.csv')
base_path='/content/drive/MyDrive/DL_Project/BCCD_train'
s=os.listdir(base_path)
d={'WBC':0, "RBC":1, "Platelets":2}

for i in s:
  xmin=[]
  xmax=[]
  ymin=[]
  ymax=[]
  c=[]
  l=train_df[train_df['filename']==i]
  for xmi in l['xmin']:
    xmin.append(xmi)
  for xma in l['xmax']:
    xmax.append(xma)
  for ymi in l['ymin']:
    ymin.append(ymi)
  for yma in l['ymax']:
    ymax.append(yma)
  for cl in l['class']:
    c.append(cl)

  file_valid=open(base_path+'/'+i[:-4]+'.txt', 'w')

  for f in range(len(xmin)):
    x=(xmin[f]+xmax[f])/2
    y=(ymin[f]+ymax[f])/2
    w=xmax[f]-xmin[f]
    h=ymax[f]-ymin[f]
    clas=c[f]
    file_valid.write(str(d[clas])+' '+str(x/640)+' '+str(y/480)+' '+str(w/640)+' '+str(h/480)+'\n')
```
